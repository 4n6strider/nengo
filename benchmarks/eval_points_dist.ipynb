# Bridson's algorithm

see http://bost.ocks.org/mike/algorithms/

```python
from scipy.special import gamma
from nengo.utils.numpy import norm

def uniform_annulus(d, rmin, rmax, rng=np.random):
    x = rng.randn(d)
    x *= rng.uniform(low=rmin**d, high=rmax**d)**(1. / d) / norm(x)
    return x
```

```python
# pick a number of samples and dimesionality
n = 1000
d = 2

# pick minimum closest distance (radius) r
rmax = 2. / (n**(1. / d) - 1)  # max radius we could accommodate assuming hyperperfect packing (impossible)
assert np.allclose((1 + rmax/2)**2, (rmax/2)**2 * n)  # full volume equals n small volumes

r = rmax / 1.2
rupper = 2 * r  # upper bound on our search disk
print r
```

```python
rng = np.random.RandomState(9)

seed = np.zeros(d)

points = np.zeros((n, d))
active_points = np.zeros(n, dtype=bool)
dists = np.zeros((n, n))
np.fill_diagonal(dists, np.inf)
points[0] = seed
active_points[0] = True
points_n = 1

while points_n < n and active_points.any():
    point_i = rng.choice(active_points.nonzero()[0])
    point = points[point_i]
    neighbours = points[:points_n][dists[point_i, :points_n] < rupper + r]
    
    for k in range(20):
        candidate = point + uniform_annulus(d, r, rupper, rng=rng)
        if norm(candidate) > 1:
            continue
            
        neighbour_dists = norm(neighbours - candidate, axis=-1)
        if (neighbour_dists < r).any():
            continue
        else:
            points[points_n] = candidate
            active_points[points_n] = True
            dists[points_n, :points_n] = dists[:points_n, points_n] = norm(points[:points_n] - candidate, axis=-1)
            points_n += 1
            break
    else:
        active_points[point_i] = False
        
print points_n
print dists[:5, :5]
```
