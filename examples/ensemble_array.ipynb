# Nengo Network: Ensemble Array

An ensemble array is a group of ensembles that each represent a dimension of the overall mutli-dimensional signal.

Functionally, ensemble arrays are similar to normal ensembles, but expose a slightly different interface. Additionally, in an ensemble array, the components of the overall signal are not related. As a result, network arrays cannot be used to compute nonlinear functions that mix the dimensions they represent. For example, you would be able to use a multi-dimensional ensemble to do multiplication, but you would not be able to use an ensemble array.

The following example shows how ensemble arrays are equivalent to multi-dimensional arrays

```python
import nengo
import numpy as np

model = nengo.Network(label='Ensemble Array')
with model:
    # Make an input node
    input_signal = nengo.Node(output=lambda t: [2*np.cos(t), 2*np.sin(t)])
    
    # Make ensembles to connect
    input_ensemble_array = nengo.networks.EnsembleArray(100, n_ensembles=2)
    plain_ensemble = nengo.Ensemble(100, dimensions=2)
    output_ensemble_array = nengo.networks.EnsembleArray(100, n_ensembles=2)
    
    # Connect the model elements, signal -> input -> normal -> output
    nengo.Connection(input_signal, input_ensemble_array.input)
    nengo.Connection(input_ensemble_array.output, plain_ensemble)
    nengo.Connection(plain_ensemble, output_ensemble_array.input)
    
    # Setup the probes for plotting
    sig_probe = nengo.Probe(input_signal)
    in_probe = nengo.Probe(input_ensemble_array.output, synapse=0.02)
    plain_probe = nengo.Probe(plain_ensemble, synapse=0.02)
    out_probe = nengo.Probe(output_ensemble_array.output, synapse=0.02)
```

```python
# Set up and run the simulator
sim = nengo.Simulator(model)
sim.run(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#Plot the results
plt.figure(figsize=(10, 8))
plt.subplot(221)
plt.plot(sim.trange(), sim.data[sig_probe])
plt.title("reference")
plt.subplot(222)
plt.plot(sim.trange(), sim.data[in_probe])
plt.title("Input Ensemble Array")
plt.subplot(223)
plt.plot(sim.trange(), sim.data[plain_probe])
plt.title("Plain Ensemble")
plt.subplot(224)
plt.plot(sim.trange(), sim.data[out_probe])
plt.title("Output Ensemble Array")
```

These plots demonstrate that the network array works very similarly to a standard N-dimensional population. However, this is not true when it comes to computing functions. Network arrays cannot be used to compute nonlinear functions that mix the dimensions they represent. Remember how we can't use an ensemble array for multiplication?

In the next example we're going to show the more subtle differences between an ensemble array and a multi-dimensional ensemble. Firstly, an emsemble array saturates across each dimension of the ouput vector, but a multi-dimensional ensemble saturates across the magnitude of the vector.

```python
import nengo
import numpy as np

model = nengo.Network(label='Ensemble Array')
with model:
    # Make an input node
    input_sig = nengo.Node(output=lambda t: [2*np.cos(t), 2*np.sin(t)])
    
    # Restrict the radius of the ensembles
    ens_array = nengo.networks.EnsembleArray(10, n_ensembles=2, radius=1)
    plain_ens = nengo.Ensemble(10, dimensions=2, radius=1)
    
    # Connect the input signal to each element
    nengo.Connection(input_sig, ens_array.input)
    nengo.Connection(input_sig, plain_ens)
    
    # Setup the probes for plotting
    sig_probe = nengo.Probe(input_sig)
    array_probe = nengo.Probe(ens_array.output, synapse=0.02)
    ens_probe = nengo.Probe(plain_ens, synapse=0.02)
```

```python
# Set up and run the simuator
sim = nengo.Simulator(model)
sim.run(10)
```

```python
#Plot the results
plt.figure(figsize=(10, 8))
plt.subplot(221)
plt.plot(sim.trange(), sim.data[sig_probe])
plt.plot(sim.trange(), sim.data[array_probe])
plt.title("Ensemble Array")
plt.subplot(222)
plt.plot(sim.trange(), sim.data[sig_probe])
plt.plot(sim.trange(), sim.data[ens_probe])
plt.title("Mutli-dimensional Ensemble")
```

Note how because there is a single ensemble processing each dimension in the ensemble array, the saturation is much more graceful.

One last thing to note about ensemble arrays is that they are quicker to build then a multi-dimensional ensemble of equal size since it is easier to calculate the weights for computing the functions.
